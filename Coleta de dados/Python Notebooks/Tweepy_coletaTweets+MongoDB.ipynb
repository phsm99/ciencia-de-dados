```python
import tweepy as tw
import pandas as pd
```

```python
auth = tw.OAuthHandler('gSZORc6y...','D4EPucTZ...')
auth.set_access_token('186857476...','k6b8S1YVqATf...')

api = tw.API(auth,wait_on_rate_limit=True)
```

```python
busca_palavras = "#vacina" + "-filter:retweets"
```

```python
tweets_filter = tw.Cursor(api.search,q=busca_palavras).items(1000)
```

```python
lista_tweets_filter = []
for tweet1 in tweets_filter:
    dados_tweet=[]
    dados_tweet.append(tweet1.created_at)
    dados_tweet.append(tweet1.user.name)
    dados_tweet.append(tweet1.user.description)
    dados_tweet.append(tweet1.user.location)
    dados_tweet.append(tweet1.text)
    
    lista_tweets_filter.append(dados_tweet)
```

```python
len(lista_tweets_filter)
```

```python
df_tweets = pd.DataFrame(lista_tweets_filter,columns=['data','nome','descricao','localidade','post'])
```
